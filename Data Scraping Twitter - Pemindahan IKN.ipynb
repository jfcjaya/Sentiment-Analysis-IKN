# Data Scraping

```python
#configuration
c = twint.Config()
# c.Search = "(pemindahan OR pindah) AND (\"IKN\" OR \"ibu kota negara\" OR \"ibu kota\") AND (\"setuju\" OR \"dukung\")"
c.Search = "(pemindahan OR pindah) AND (\"ikn\" OR \"ibu kota negara\" OR \"ibu kota\") AND (\"tolak\" OR \"tidak setuju\")"
c.Count = True
c.Store_object = True
c.Lang = "id"
c.Links = "exclude"
c.Pandas = True
c.Retries_count = 100
c.Filter_retweets = True

c.Since = "2021-1-01"
c.Until = "2021-3-28"

#running search
twint.run.Search(c)
```

```python
df = twint.storage.panda.Tweets_df
df.shape
```

```python
df.sort_values('tweet', inplace = True)
df.drop_duplicates(subset=['tweet'], inplace=True)
df = df.reset_index(drop = True)

df.shape
```

```python
import pandas as pd
import numpy as np

df['sentiment'] = np.nan
df.head()
```

```python
columns = ['id', 'tweet', 'sentiment']
df = df[columns]
df.shape
```

```python
import json
result = df.to_json(orient = "index")
parsed = json.loads(result)
json.dumps(parsed, indent = 4)
```

# Temporary Data Combiner

Bagian ini digunakan untuk menggabungkan hasil scraping dari kumpulan data yang telah discrap. Terdapat 16 file terpisah dengan nama 1.json hingga 16.json.

```python
df = pd.read_json("1.json", orient = "index")

for idx in range(2, 17):
    file_name = str(idx) + ".json"
    temp = pd.read_json(file_name, orient = "index")
    df = df.append(temp)
```

```python
df.shape
```

```python
df.sort_values('tweet', inplace = True)
df.drop_duplicates(subset=['tweet'], inplace=True)
df = df.reset_index(drop = True)

df.shape
```

```python
import json
result = df.to_json(orient = "index")
parsed = json.loads(result)
json.dumps(parsed, indent = 4)
```
